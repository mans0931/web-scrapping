```python
##importing libraries
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup 
```

```python
##upload html
my_url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
soup = BeautifulSoup(page_html, 'html')
```

```python
table = soup.find_all('table')[0]
```

```python
world_titles = soup.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

# Method 3: Loop with Removal
world_table_titles_filtered = world_table_titles.copy()  # Create a copy

# Iterate backwards to avoid index issues
for i in range(len(world_table_titles_filtered) - 1, -1, -1):
    if not world_table_titles_filtered[i]:
        del world_table_titles_filtered[i]

print(world_table_titles_filtered)  # This will print the filtered list
```

```python
world_table_titles_1 = world_table_titles_filtered[1:8]
world_table_titles_1
```

```python
world_table_titles_2 = world_table_titles_filtered[8:11:2]
world_table_titles_2
```

```python
combined = world_table_titles_1 + world_table_titles_2
combined
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(columns = combined)
df
```
